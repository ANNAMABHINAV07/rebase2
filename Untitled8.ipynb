```python
def is_valid_phone_number(phone_number):
    # Removing spaces for easier processing
    phone_number = phone_number.replace(" ", "")
    
    # Check if it starts with +91, 0, or a digit between 7 and 9
    if phone_number.startswith("+91"):
        # Remove the +91 and check the remaining digits
        phone_number = phone_number[3:]
    elif phone_number.startswith("0"):
        # Remove the 0 and check the remaining digits
        phone_number = phone_number[1:]
    elif phone_number[0] in "789":
        # Valid if it starts with 7, 8, or 9
        pass
    else:
        # Invalid if none of the above conditions are met
        return False
    
    # After stripping, the remaining phone number should be exactly 10 digits long
    if len(phone_number) != 10 or not phone_number.isdigit():
        return False
    
    return True
def main():
    # Read the number of phone numbers
    n = int(input("Enter the number of phone numbers: "))
    
    # Iterate over the phone numbers
    for _ in range(n):
        phone_number = input().strip()
        # Validate and print result
        print(is_valid_phone_number(phone_number))

if __name__ == "__main__":
    main()










```
